```python
df = df.drop(['CouponId', 'ReferenceId', 'BookingCode','Notes','Description',
                'RefundDescription', 'CancelReason', 'UtmSource', 'Email',
                'GuestPaymentTime','HostDeadLineTime', 'PaymentGatewayId',
                'CompensationDescription','LastFourCharactersOfCard','FailedReason'], axis = 1)
```

```python
df['FullName'] = df.FullName.fillna('0')
```

```python
df['GuestId'] = df.GuestId.fillna(df.GuestId.mean())
```

```python
df['NationalCode'] = df.NationalCode.fillna('0')
```

```python
df['UpdatedBy'] = df.UpdatedBy.fillna(df.UpdatedBy.mean())
```

```python
df['UpdatedOn'] = df.UpdatedOn.fillna('0')
```

```python
df['CellPhone'] = df.CellPhone.fillna(df.CellPhone.mean())
```
